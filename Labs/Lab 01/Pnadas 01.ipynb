```python
people = {
    "name" : ["Rauf", "Arif", "Maaz", "Hadeed", "Mujahid", "Mohid"],
    "age" : [52, 51, 26, 22, 18, 17],
    "address": ["Lahore", "Karachi", "Lahore", "Islamabad", "Kakul", "Karachi"],
    "cell" : ["321-123", "320-431", "321-478", "324-446", "321-967", "320-678"],
    "bg": ["B+", "A-", "B+", "O-", "A-", "B+"]
}

df_people = pd.DataFrame(data=people, index=['MS01', 'MS02', 'MS03', 'MS04', 'MS05', 'MS06'])
df_people
```

```python
df_temp = pd.DataFrame({
    'city': ['Lahore', 'Muree', 'Peshawer', 'Sialkot'],
    'temperature' : [39, 14, 29, 32]
})
df_temp
```

```python
df_hum = pd.DataFrame({
    'city': [ 'Karachi', 'Lahore', 'Peshawer', 'Lahore', 'Muree'],
    'humidity' : [76, 95, 72, 70, 75]
})
df_hum
```

```python
d1 = pd.merge(df_temp, df_hum, how='inner')
d1
```

```python
d2 = pd.merge(df_temp, df_hum, on='city', how='outer', indicator=True)
d2
```

```python
d3 = pd.merge(df_temp, df_hum, on='city', how='left', indicator=True)
d3
```

```python
d3 = pd.merge(df_temp, df_hum, on='city', how='right', indicator=True)
d3
```

```python
df1 = pd.DataFrame({
    'city': [ 'Lahore', 'Karachi', 'Peshawer'],
    'temperature' : [35, 39, 33],
    'humidity' : [76, 95, 72]
})

df1
```

```python
df2 = pd.DataFrame({
    'city': [ 'Karachi', 'Peshawer', 'Islamabad'],
    'temperature' : [41, 44, 47],
    'humidity' : [88, 99, 79]
})
df2
```

```python
d3 = pd.merge(df1, df2, on='city', how='inner', )
d3
```

```python
d3 = pd.merge(df1, df2, on='city', how='inner', suffixes=('_left','_right'))
d3
```
