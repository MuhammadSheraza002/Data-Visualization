# 5. Enhance plot with annotations: (4)
- Add an annotation on a random point in plot

```python
plt.plot(x, y, marker='s', color='blue')
plt.grid(True)
plt.title("Plot of y and x values")
plt.xlabel("X-values")
plt.ylabel("Y-values")
plt.annotate('Maximum Point', xy=(3,7), xytext=(2.5,3),
             arrowprops=dict(facecolor='green',shrink=0.1),
             fontsize=20, color='black')

plt.xlim(0,6)
plt.ylim(0, 8);
```

# 6. Subplots and multiple plots: (4)
Create 2 Subplots
+ Bar Plot
+ Scatter Plot

```python
fig, (plot1, plot2) = plt.subplots(1, 2, figsize=(10, 4))


plot1.bar(x, y, color='blue')
plot1.grid(True)
plot1.set_title("Bar Plot")
plot1.set_xlabel("X-values")
plot1.set_ylabel("Y-values")
plot1.set_xlim(0,6)
plot1.set_ylim(0,8)


plot2.scatter(x, y, color='red')
plot2.grid(True)
plot2.set_title("Scatter Plot")
plot2.set_xlabel("X-values")
plot2.set_ylabel("Y-values")

plot2.set_xlim(0,6)
plot2.set_ylim(0,8)


plt.tight_layout()


plt.show()
```

#### Apply Color Maps to Data
Generate a 10x10 heatmap of random data using the 'coolwarm' colormap and display the color bar with labeled data values.


```python
data = np.random.rand(10, 10)

heatmap = plt.imshow(data, cmap='coolwarm',interpolation='nearest')

cbar = plt.colorbar(heatmap)
cbar.set_label('Data Values')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Heatmap with Coolwarm Colormap');
```

# 9. Layout and composition: (4)

Explore the impact of different plot sizes and aspect ratios by creating a small square plot and a larger wide plot using the 'scatter' function.
- Small = 5x5
- Large = 10x5

```python

plt.figure(figsize=(5, 5))
plt.scatter(x, y, color='blue', marker='o')
plt.title('Small Square Plot (5x5)')


plt.figure(figsize=(10, 5))
plt.scatter(x, y, color='green', marker='s')
plt.title('Large Wide Plot (10x5)')


plt.show()
```

Modify the legend placement to the upper left and adjust the positioning of the annotation for the point (2, 5) in a scatter plot

```python

plt.scatter(x, y, color='blue', marker='o', label='Random Data')


plt.legend(loc='upper left')


plt.annotate('Point (2, 5)', xy=(2, 5), xytext=(2.5, 5.5),
             arrowprops=dict(facecolor='black', shrink=0.05),
             fontsize=10, color='green')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Legend and Annotation')
plt.show()
```

Enhance the use of whitespace in a plot by adjusting the spacing around the plot area and creating a scatter plot with improved whitespace.

```python
plt.scatter(x, y, color='blue', marker='o', label='Random Data')
plt.subplots_adjust(left=0.1, right=0.9, top=0.9, bottom=0.1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Improved Whitespace')


plt.show()
```

Utilize color gradients to represent data using a specified colormap, for example, 'YlGnBu', to create a scatter plot with a color bar.
